```python
import qiskit
from qiskit.providers.aqt import AQT
from qiskit.visualization import plot_histogram
aqt = AQT.enable_account('2633b2d14dae4d9da7c9df2503857334')
```

```python
ideal = aqt.get_backend('aqt_qasm_simulator')
noise = aqt.get_backend('aqt_qasm_simulator_noise_1')
```

```python
qc = qiskit.QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0,1], [0,1])

qc.draw(output='mpl')
```

```python
res_ideal = qiskit.execute(qc, ideal, shots=200).result()
plot_histogram(res_ideal.get_counts(qc))
```

```python
res_noise = qiskit.execute(qc, noise, shots=200).result()
plot_histogram(res_noise.get_counts(qc))
```
